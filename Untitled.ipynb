```python
import pandas as pd
```

```python
nov = "202010-citibike-tripdata.csv"
dec = "202011-citibike-tripdata.csv"
jan = "202011-citibike-tripdata.csv"
feb = "202011-citibike-tripdata.csv"

nov_df = pd.read_csv(nov)
dec_df = pd.read_csv(dec)
jan_df = pd.read_csv(jan)
feb_df = pd.read_csv(feb)
```

```python
labels = [ '01 November', '02 December']
frames = [nov_df, dec_df]



for x in range(2):
    temp_list = []
    for i in range(len(frames[x])):
        temp_list.append(labels[x])
    
    (frames[x])['month'] = temp_list
```

```python
frames_mrg = [nov_df, dec_df]

df_20 = pd.concat(frames_mrg)
```

```python
year = []
for i in range(len(df_20["bikeid"])):
    year.append(2020)
    
df_20["year"] = year
```

```python
labels2 = ['03 January', '04 February']
frames2 = [jan_df, feb_df]


for x in range(2):
    temp_list2 = []
    for i in range(len(frames2[x])):
        temp_list2.append(labels2[x])
    
    (frames2[x])['month'] = temp_list2
```

```python
frames_mrg2 = [jan_df, feb_df]

df_21 = pd.concat(frames_mrg2)
```

```python
year2 = []
for i in range(len(df_21["bikeid"])):
    year2.append(2021)
    
df_21["year"] = year2
```

```python
year_frames = [df_20, df_21]
combined_csv = pd.concat(year_frames)
```

```python
gender_list = combined_csv["gender"].tolist()

male_list = [] 
female_list = []

for x in gender_list:
    if x == 0:
        male_list.append(0) 
        female_list.append(0) 
    elif x == 1:
        male_list.append(1) 
        female_list.append(0)     
    elif x == 2:
        male_list.append(0) 
        female_list.append(1)
```

```python
combined_csv["male"] = male_list
```

```python
combined_csv["female"] = female_list
```

```python
combined_csv = combined_csv.drop(['gender'], 1)
```

```python
combined_csv['tripduration(minutes)'] = (combined_csv['tripduration'] / 60)
```

```python
combined_csv['tripduration(hours)'] = (combined_csv['tripduration(minutes)'] / 60)
```

```python
time_combo = []
for y in range(len(combined_csv)):
    time_combo.append((combined_csv['month'].iloc[y]) + " " + str(combined_csv['year'].iloc[y]))
```

```python
combined_csv["datetime"] = time_combo
```

```python
combined_csv = combined_csv.drop(['month'], 1)
```

```python
combined_csv = combined_csv[combined_csv['birth year'] >= 1940 ]
```

```python
combined_csv = combined_csv[combined_csv['tripduration'] <= 43200]
```

```python
combined_csv = combined_csv[combined_csv['tripduration'] >= 60]
```
